```python
import numpy as np
```

构造矩阵$A$和向量$b$

```python
a = [[1 + (299 + I) * int(I == J) for I in range(300)] for J in range(300)]
b = [(299 + I) * I + 45150 - I for I in range(1, 301)]
a = np.array(a, dtype='float')
b = np.array(b, dtype='float')
```

```python
display(a)
```

```python
display(b)
```

应用列主元消去法变换$A$和$b$

```python
for K in range(300):
    # 换行
    m = a[:, K].argmax()
    temp = a[m], b[m]
    a[m], b[m] = a[K], b[K]
    a[K], b[K] = temp
    for I in range(K + 1, 300):
        # 消去
        l = a[I, K] / a[K, K]
        a[I] = a[I] - a[K] * l
        b[I] = b[I] - b[K] * l
```

回代计算$x$

```python
x = np.zeros_like(b)
x[-1] = b[-1] / a[-1, -1]
for I in range(298, -1, -1):
    x[I] = (b[I] - a[I].dot(x)) / a[I, I]
```
