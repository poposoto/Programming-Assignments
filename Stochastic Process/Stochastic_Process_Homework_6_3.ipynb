```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn
random = np.random.RandomState(2202566)
seaborn.set()
```

```python
fig, ax = plt.subplots(figsize=(24, 12))
t = np.arange(10001) / 10000
for I in range(20):
    w = [0]
    for J in t[1:]:
        v = random.normal()
        w.append(w[-1] + 0.01 * v)
    w = np.array(w) # Brownian motion
    z = w - t * (w[-1] - 1) # Brownian bridge
    ax.plot(t, z)
ax.tick_params(labelsize=30)
```
